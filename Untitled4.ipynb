```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f0(a, b):
    #0,1 - simul
    output = a*(1-(1+b)/2)-a*(1-a)-(a**2)/(2*(1+b))
    return output
# 
def f1(a, b):
    #01-simul
    return a*(1-a*(1-(1+b)/2)) - a + a**2/(2*(1+b))

a = np.linspace(0, 1, 30)
b = np.linspace(0, 5, 30)

A, B = np.meshgrid(a, b)
Z0 = f0(A, B)
Z1 = f1(A, B)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(A, B, Z0, 50, cmap='binary')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('z');
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(A, B, Z1, 50, cmap='binary')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('z');
```
